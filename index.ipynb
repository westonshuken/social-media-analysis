```python
X = df[['Voice', 'Photo Face', 'Chinese Culture', 'Romance', 'US Culture/DOL/Journal']]
y = df['LikeCommentMetric']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=88)

rf_regressor = RandomForestRegressor(random_state=88)

# Fit the model on training data
rf_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Check feature importance
importances = rf_regressor.feature_importances_
feature_names = X.columns

# Print feature importances
for feature, importance in zip(feature_names, importances):
    print(f'{feature}: {importance:.4f}')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
print(model.intercept_)
```

```python
coeff_parameter = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_parameter
```

```python
predictions = model.predict(X_test)
predictions
```

```python
import statsmodels.api as sm
X_train_Sm= sm.add_constant(X_train)
ls=sm.OLS(y_train,X_train_Sm).fit()
print(ls.summary())
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score


model = DecisionTreeRegressor(max_depth=3)
model.fit(X_train, y_train)

# Predict values using the testing data
y_pred = model.predict(X_test)

# Get the feature importance
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': model.feature_importances_
})

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate R-squared (R2)
r2 = r2_score(y_test, y_pred)

# Print results
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2): {r2}")
print(feature_importance)
```

```python
# Plotting the feature importance
plt.figure(figsize=(8,6))
plt.barh(feature_importance['Feature'], feature_importance['Importance'])
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Feature Importance in Decision Tree')
plt.show()
```

```python
from scipy.stats import shapiro

# Shapiro-Wilk test for normality
stat, p_value = shapiro(df['Likes'])

print(f"Shapiro-Wilk Test Statistic: {stat}")
print(f"P-value: {p_value}")
```

```python
from scipy.stats import kstest

# Kolmogorov-Smirnov test
stat, p_value = kstest(df['Likes'], 'norm')

print(f"Kolmogorov-Smirnov Test Statistic: {stat}")
print(f"P-value: {p_value}")
```

```python
for feature in df.columns[8:]:
    # Split data into two groups based on the binary feature 'Voice'
    group_1 = df[df[feature] == 1]['Likes']
    group_0 = df[df[feature] == 0]['Likes']

    # Perform two-sample t-test
    t_stat, p_value = stats.ttest_ind(group_1, group_0, equal_var=False)

    # Print the results
    print(f"T-statistic {feature}: {t_stat}")
    print(f"P-value {feature}: {round(p_value,4)}")
    print(" ")
```

```python
from scipy import stats

for metric in ['Likes', 'Comments', 'Boost #']:
    # Split data into two groups based on the binary feature 'Voice'
    group_1 = df[df['Voice'] == 1][metric]
    group_0 = df[df['Voice'] == 0][metric]

    # Perform two-sample t-test
    t_stat, p_value = stats.ttest_ind(group_1, group_0, equal_var=False)

    # Print the results
    print(f"T-statistic {metric}: {t_stat}")
    print(f"P-value {metric}: {p_value}")
    print(" ")
```

```python
# Perform Mann-Whitney U Test

group_1 = df[df['Voice'] == 1]['Likes']
group_0 = df[df['Voice'] == 0]['Likes']

u_stat, p_value_u = stats.mannwhitneyu(group_1, group_0)

# Print the results
print(f"U-statistic: {u_stat}")
print(f"P-value (Mann-Whitney): {p_value_u}")
```

```python
from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U test and get the statistic
u_stat, p_value = mannwhitneyu(group_1, group_0)
n_1 = len(group_1)
n_0 = len(group_0)
N = n_1 + n_0

# Assuming you calculate Z (some libraries give Z directly)
z = (u_stat - (n_1 * n_0 / 2)) / np.sqrt(n_1 * n_0 * (n_1 + n_0 + 1) / 12)

# Calculate r
r = z / np.sqrt(N)
print(f"Effect size (r): {r}")
```
